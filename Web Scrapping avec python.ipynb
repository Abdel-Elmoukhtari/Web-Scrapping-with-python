```python
#import Les packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Saisir l'url
url="http://feeds.bbci.co.uk/news/rss.xml"
```

```python
#import le code de la page
#r = requests.get('http://feeds.bbci.co.uk/news/rss.xml')
reponse=requests.get(url)
soup=BeautifulSoup(reponse.text,'html.parser')
```

```python
items=soup.findAll('item')
```

```python
item=items[0]
```

```python
item.title.text
```

```python
item.description.text
```

```python
news_items=[]
```

```python
for i in items:
    news_i={}
    news_i['title']=i.title.text
    news_i['description']=i.description.text
    news_i['pubdate']=i.pubdate.text
    news_items.append(news_i)
```

```python
news_items[0]
```

```python
#convertir la liste en base de donnees
df=pd.DataFrame(news_items,columns=['title','description','pubdate'])
```

```python
df.head()
```

```python
#genere le fichier csv
df.to_csv('web_scrapping.csv',index=False,encoding='utf-8')
print('la base de donneer et creier')
```

# Initiation au Web Scraping avec Python: Extraire des articles et les mettre en base de données

```python
!pip install newspaper3k
from newspaper import Article
!pip install nltk
import nltk
import pandas as pd
```

```python
#specifier les article qui m'interesst
urls=["https://www.20minutes.fr/societe/4027134-20230311-stages-fourre-tout-efficacite-lutter-contre-sexisme-reste-prouver","https://www.20minutes.fr/sport/4027440-20230311-ski-alpin-87-reine-shiffrin-bat-record-nombre-victoires-coupe-monde","https://www.20minutes.fr/monde/ukraine/4027411-20230311-guerre-ukraine-direct-centaine-attaques-russes-repoussees-region-donetsk"]
```

```python
def url_text(url):
    article=Article(url)
    article.download()
    article.parse()
    titre=article.title
    text=article.text
    auteur=article.authors
    print(url)
    return titre,auteur,text

texts=[url_text(url) for url in urls ]
```

```python
texts[0]
```

```python
colonnes_nom=['titre','auteur','text']
df=pd.DataFrame(data=texts,columns=colonnes_nom)
```
